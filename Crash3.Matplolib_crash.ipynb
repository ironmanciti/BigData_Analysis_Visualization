````python
from matplotlib import gridspec

plt.figure()
gs = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gs[0, 1:])
side_histogram = plt.subplot(gs[1:, 0])
lower_right = plt.subplot(gs[1:, 1:])

X = np.random.random(1000)
Y = np.random.randn(1000)

lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal')

side_histogram.set_ylabel('Y')
lower_right.set_xlabel("X")

plt.tight_layout()
````

````python
gridsize = (3, 2)
fig = plt.figure(figsize=(10, 6))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))

#건축연수와 인구에 따른 주택 가격
ax1.set_title("House Price", fontsize=10)
sctr = ax1.scatter(age, pop, c=y, cmap="RdYlGn")
plt.colorbar(sctr, ax=ax1, format='$%d')    
ax1.set_yscale('log')
ax2.hist(age, bins='auto');
ax3.hist(pop, bins='auto', log=True);

ax2.text(.5, .5, 'House Age Histogram', transform=ax2.transAxes)
ax3.text(.5, .5, 'Demography Histogram (log scale)', transform=ax3.transAxes);

fig.suptitle("GridSpec plot");
````

## meshgrid / contour (3 차원 자료의 2 차원 표현)

- 입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우에는 3차원 자료가 됨   


- 3차원 자료를 2차원 평면에 시각화하는 방법 


    - 색 (color) 로 구분  
    - 등고선(contour) 으로 표시 -  contour 혹은 contourf 명령 사용 (contour는 등고선만 표시하고 contourf는 색깔로 표시)
    - 입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 함
    

- meshgrid : coordinate vector 를 이용하여 coordinate matrix 구성

## The Dataset: Immigration to Canada from 1980 to 2013 <a id="2"></a>

Dataset Source: [International migration flows to and from selected countries - The 2015 revision](http://www.un.org/en/development/desa/population/migration/data/empirical2/migrationflows.shtml).

이 데이터 세트에는 목적지 국가에서 기록한 국제 이민자의 흐름에 대한 연간 데이터가 포함되어 있습니다. 자료는 출생지, 시민권 또는 외국인과 국민 모두를위한 이전 / 다음 거주지에 따라 유입과 유출을 모두 나타냅니다. 현재 버전은 45 개국에 관한 데이터를 제공합니다.

이 실습에서는 캐나다 이민 데이터에 중점을 둘 것입니다.

````python
# !pip install openpyxl
````

````python
import pandas as pd
import matplotlib.patches as mpatches # needed for waffle Charts

df_can = pd.read_excel('Canada.xlsx', engine='openpyxl',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

# 데이터 세트를 정리하여 불필요한 열(예 : REG)을 제거
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)

# 이해하기 쉽게 열 이름을 변경합니다.
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)

# 국가 이름을 색인으로 설정하면 .loc 메소드를 사용하여 국가를 빠르게 찾는 데 유용합니다.
df_can.set_index('Country', inplace=True)

# 이 노트북에서 사용할 연도 - 나중에 플로팅하는 데 유용합니다.
years = list(map(str, range(1980, 2014)))

# 일관성을 위해 모든 열 레이블을 문자열 유형으로 만듭니다.
df_can.columns = list(map(str, df_can.columns))
condition = df_can['Continent'] == 'Asia'
condition = (df_can['Continent'] == 'Asia') & (df_can['Region'] == 'Southern Asia')

# add total column
df_can['Total'] = df_can.sum(axis=1)

print ('data dimensions:', df_can.shape)
````

# Waffle Charts 

`waffle chart`는 일반적으로 목표를 향한 진행 상황을 표시하기 위해 만드는 흥미로운 시각화입니다. 일반적으로 Excel 대시 보드와 같이 주로 셀로 구성된 시각적 개체에 흥미로운 시각화 특성을 추가하려고 할 때 효과적인 옵션입니다.

덴마크, 노르웨이 및 스웨덴에 대한 사례 연구를 살펴 보겠습니다.

````python
df_dsn = df_can.loc[['Denmark', 'Norway', 'Sweden'], :]

df_dsn
````

** 1 단계 ** 와플 차트를 만드는 첫 번째 단계는 합계에 대한 각 범주의 비율을 결정하는 것입니다.

````python
# 합계에 대한 각 범주의 비율을 계산
total_values = sum(df_dsn['Total'])
category_proportions = [(float(value) / total_values) for value in df_dsn['Total']]

# print out proportions
for i, proportion in enumerate(category_proportions):
    print (df_dsn.index.values[i] + ': ' + str(proportion))
````

** 2 단계 ** 두 번째 단계는 '와플' 차트의 전체 크기를 정의하는 것입니다.

````python
width = 40 # width of chart
height = 10 # height of chart

total_num_tiles = width * height # total number of tiles

print ('Total number of tiles is ', total_num_tiles)
````

** 3 단계 ** 세 번째 단계는 각 카테고리의 비율을 사용하여 각 타일 수를 결정하는 것입니다.

````python
# 각 카테고리의 타일 수를 계산
tiles_per_category = [round(proportion * total_num_tiles) for proportion in category_proportions]
print(tiles_per_category)

# 카테고리 별 타일 수 출력
for i, tiles in enumerate(tiles_per_category):
    print (df_dsn.index.values[i] + ': ' + str(tiles))
````

계산 된 비율에 따르면 덴마크는 '와플'차트에서 129 개의 타일을, 노르웨이는 77 개의 타일을, 스웨덴은 194 개의 타일을 차지할 것입니다.


** 4 단계 ** 네 번째 단계는 '와플'차트와 유사한 행렬을 만들어 채우는 것입니다.

````python
# initialize the waffle chart as an empty matrix
waffle_chart = np.zeros((height, width))

# define indices to loop through waffle chart
category_index = 0
tile_index = 0

# populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1

        #현재 카테고리에 채워진 타일 수가 할당된 해당 타일과 같은 경우
        if tile_index > sum(tiles_per_category[0:category_index]):
            # ...proceed to the next category
            category_index += 1       
            
        # set the class value to an integer, which increases with class
        waffle_chart[row, col] = category_index
        
print(waffle_chart.shape)
print()
print(waffle_chart)
````

예상대로 매트릭스는 세 가지 범주로 구성되며 각 범주의 총 인스턴스 수는 각 범주에 할당된 총 타일 수와 일치합니다.

** 5 단계. ** `waffle` 차트 매트릭스를 visual에 매핑합니다.

````python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
````

** 6 단계 ** 차트를 예쁘게합니다.  


** 7 단계 ** 범례를 만들고 차트에 추가합니다.

````python
# 6단계
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])

# 7단계

# 차트와 범례 사이의 색 구성표를 일치시키기 위해 개별 범주의 누적 합계를 계산합니다.
values_cumsum = np.cumsum(df_dsn['Total'])
total_values = values_cumsum[len(values_cumsum) - 1]

# 범례 생성
legend_handles = []
for i, category in enumerate(df_dsn.index.values):
    label_str = category + ' (' + str(df_dsn['Total'][i]) + ')'
    color_val = colormap(float(values_cumsum[i])/total_values)
    legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

# add legend to chart
plt.legend(handles=legend_handles,
           loc='lower center', 
           ncol=len(df_dsn.index.values),
           bbox_to_anchor=(0., -0.2, 0.95, .1)
          )
````

## Word Cloud 

`Word Cloud` (텍스트 클라우드 또는 태그 클라우드라고도 함)는 간단한 방식으로 작동합니다. 텍스트 데이터 소스 (예 : 음성, 블로그 게시물 또는 데이터베이스)에 특정 단어가 더 많이 나타날수록 더 크고 굵게 표시됩니다. 

`Word Cloud`는 일반적으로 텍스트 데이터의 고급 분석 및 시각화를 수행하는 데 사용됩니다. *이상한 나라의 엘리스*라는 단편 소설을 분석해 보겠습니다.

````python
from wordcloud import WordCloud, STOPWORDS

# open the file and read it into a variable alice_novel
alice_novel = open('alice_novel.txt', 'r').read()
````

다음으로 `wordcloud`에서 가져온 불용어를 사용하겠습니다. *set* 함수를 사용하여 중복 불용어를 제거합니다.

````python
stopwords = set(STOPWORDS)
````

단어 구름 개체를 만들고 단어 구름을 생성합니다. 단순화를 위해 소설의 처음 2000 개 단어 만 사용하여 단어 구름을 생성 해 보겠습니다.

````python
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)
````

````python
# display the word cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
````

소설의 처음 2000 개 단어에서 가장 일반적인 단어는 **Alice**, **said**, **little**, **Queen** 등 입니다. **said**은 실제로 유용한 단어가 아닙니다. 이제 불용어에 추가하고 클라우드를 다시 생성 해 보겠습니다.

자주 사용하지 않는 단어를 조금 더 잘 볼 수 있도록 클라우드 크기를 조정 해 보겠습니다.

````python
stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
````

`word_cloud` 패키지로 구현할 수있는 또 다른 멋진 기능은 모든 모양의 마스크에 단어를 겹쳐 놓는 것입니다. 

````python
from PIL import Image
alice_mask = np.array(Image.open('alice_mask.png'))
````

````python
fig = plt.figure()
fig.set_figwidth(4) # set width
fig.set_figheight(6) # set height

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
````

````python
# instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

# generate the word cloud
alice_wc.generate(alice_novel)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
````

### Data Science 분야에 대한 관심도 조사 결과를 시각화

````python
df_survey = pd.read_csv('Topic_Survey_Assignment.csv', index_col=0)
df_survey
````

### 1 단계 - 관심도 높은 순서로 정렬

````python
df_survey.sort_values('Very interested')
df_survey
````

### 2 단계 - 응답한 숫자를 전체 응답 숫자에 대한 비율로 전환

````python
row_sum = df_survey.sum(axis=1)
row_sum
````

````python
df_survey = round(df_survey.divide(row_sum, axis=0) * 100, 2)
df_survey
````

### 3 단계 - 조사 분야별로 bar chart 작성하여 비교

````python
categories = df_survey.index
categories
````

````python
fig, ax = plt.subplots(figsize=(16, 6))

index = np.arange(len(df_survey))

very_interested = df_survey.iloc[:, 0].values
somewhat_interested = df_survey.iloc[:, 1].values
not_interested = df_survey.iloc[:, 2].values

ax.bar(index, very_interested, width=0.3, color='red', label="Very Interested")

ax.bar(index+0.3, somewhat_interested, width=0.3, color='green', 
       label="somewhat_interested", tick_label=categories)                  

ax.bar(index+0.6, not_interested, width=0.3, color='blue', label="not_interested")

ax.set_title("Percentage of the Respondent's interest in Data Science Area", fontsize=16)
ax.set_xticklabels(categories, rotation=90)

plt.legend(fontsize=11)
````
